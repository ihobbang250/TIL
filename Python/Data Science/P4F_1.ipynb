## Plotly
- 시각화를 웹과 연동하여 그릴수 있다

```python
import pandas as pd
import cufflinks as cf
import plotly.offline as plyo
import numpy as np

plyo.init_notebook_mode(connected=True) #Notemode -> Offline rendering
```

```python
df = pd.read_csv('BTC-USD.csv', index_col=0, parse_dates=True)

raw = df[['Open', 'High', 'Low', 'Close']]
raw = raw.iloc[-100:]
raw.head()
```

```python
# Cufflinks Package
qf = cf.QuantFig(raw, title='BTC', legend='top', name='BTC/USD')

plyo.iplot(
    qf.iplot(asFigure=True),
    filename='qf_01'
)
```

```python
qf.add_bollinger_bands(periods=10, boll_std=2)

plyo.iplot(
    qf.iplot(asFigure=True),
    filename='qf_02'
)
```

```python
qf.add_rsi(periods=5,
           showbands=False)

plyo.iplot(
    qf.iplot(asFigure=True),
    filename='qf_03'
)
```
